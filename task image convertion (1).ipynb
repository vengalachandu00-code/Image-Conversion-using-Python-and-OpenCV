```python
#class img_conversion
#instance attribute - path of the image
#method1 --> displaying the image in color mode
#method2 --> displaying the image in gray scale mode
#method3 --> displaying the image in black and white mode (100-180)
```

```python
import cv2
```

```python
class img_conversion:
    def __init__(self,img_path):
        self.img_path=img_path

    def color(self):
        img=cv2.imread(self.img_path)
        cv2.imshow("image",img)
        cv2.waitKey(5000)
        cv2.destroyAllWindows()
    def gray(self):
        img=cv2.imread(self.img_path,0)
        cv2.imshow("image",img)
        cv2.waitKey(5000)
        cv2.destroyAllWindows()
    def white(self,a,b):
        img2 = cv2.threshold(cv2.imread(self.img_path,0), a, b, cv2.THRESH_BINARY)[1]
        cv2.imshow("charan",img2)
        cv2.waitKey(5000)
        cv2.destroyAllWindows()
```

```python
obj = img_conversion(r"C:\Users\chandu\Downloads\a.jpg")
obj.color()
obj.gray()
obj.white(100,180)
```

```python
class img_conversion:
    def __init__(self,img_path):
        self.img_path=img_path
    def color(self):
        img=cv2.imread(self.img_path)
        cv2.imshow("image",img)
        cv2.waitKey(5000)
        cv2.destroyAllWindows()
        
```

```python
obj1=img_conversion(r"C:\Users\chandu\Downloads\a.jpg")
print(obj1.color())
```

```python
class grey(img_conversion):
    
    def grey_color(self):
        img=cv2.imread(self.img_path,0)
        cv2.imshow("image",img)
        cv2.waitKey(5000)
        cv2.destroyAllWindows()
    
```

```python
x=grey(r"C:\Users\chandu\Downloads\a.jpg")
x.grey_color()
```

```python
class black(grey):
    def __init__(self,img_path,a,b):
        super().__init__(img_path)
        self.a = a 
        self.b = b
    def white(self):
        img2 = cv2.threshold(cv2.imread(self.img_path,0), self.a, self.b, cv2.THRESH_BINARY)[1]
        cv2.imshow("charan",img2)
        cv2.waitKey(5000)
        cv2.destroyAllWindows()
```

```python
y=black(r"C:\Users\chandu\Downloads\a.jpg",a=100,b=180)
y.white()
```
